```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parametry rozkładu
mu = 100  # Prawdziwa średnia
b = 10  # Skala (równoważnik odchylenia standardowego dla rozkładu normalnego)
n = 100

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, num_samples in enumerate([100, 1000, 10000, 100000, 500000, 1000000]):
    # Generowanie próbek i estymacja średniej
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.laplace(mu, b, n)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Wizualizacja rozkładu estymatorów średniej
    ax = axes[i // 3, i % 3]
    sns.histplot(sample_means, kde=True, stat="density", ax=ax)
    ax.set_title(f"{num_samples} próbek z rozkładu Laplace'a(mu, b)")
    ax.set_xlabel("Średnia próbkowa")
    ax.set_ylabel("Gęstość")
    ax.axvline(x=mu, color='r', linestyle='--', label=f'Prawdziwa średnia: {mu}')
    ax.legend()

plt.tight_layout()
plt.show()
```
