```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parametry rozkadu
mu = 100  # Prawdziwa rednia
b = 10  # Skala (rwnowanik odchylenia standardowego dla rozkadu normalnego)
n = 100

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, num_samples in enumerate([100, 1000, 10000, 100000, 500000, 1000000]):
    # Generowanie prbek i estymacja redniej
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.laplace(mu, b, n)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Wizualizacja rozkadu estymatorw redniej
    ax = axes[i // 3, i % 3]
    sns.histplot(sample_means, kde=True, stat="density", ax=ax)
    ax.set_title(f"{num_samples} prbek z rozkadu Laplace'a(mu, b)")
    ax.set_xlabel("rednia prbkowa")
    ax.set_ylabel("Gsto")
    ax.axvline(x=mu, color='r', linestyle='--', label=f'Prawdziwa rednia: {mu}')
    ax.legend()

plt.tight_layout()
plt.show()
```
